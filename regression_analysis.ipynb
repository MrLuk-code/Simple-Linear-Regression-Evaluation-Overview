# Simple Linear Regression – Marketing ROI Analysis
### Project Overview
This notebook provides a complete data science workflow using Python and `statsmodels` to build a Simple Linear Regression model. We analyze a marketing dataset to identify which advertising channel (TV, Radio, or Social Media) yields the highest return on investment (ROI) for Sales. 

### Project Tasks Completed:
1. Load and clean the dataset (handling missing values).
2. Perform Exploratory Data Analysis (EDA) with visualizations.
3. Identify the independent variable most correlated with Sales.
4. Build an Ordinary Least Squares (OLS) regression model.
5. Create diagnostic plots to test core assumptions: Linearity, Normality, and Homoscedasticity.
6. Interpret model statistics ($R^2$, coefficients, p-values).
7. Formulate data-driven budget allocation recommendations.

## 1. Setup and Library Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import scipy.stats as stats

# Configure plotting aesthetics
sns.set_theme(style="whitegrid")
plt.rcParams["figure.figsize"] = (12, 5)
print("Libraries successfully imported.")
```

## 2. Load and Clean the Dataset
We inspect the data for completeness and view key summary statistics.

```python
# Load the marketing data
filepath = "marketing_and_sale_data_evaulate_Ir.csv"
df = pd.read_csv(filepath)

# Display the first few rows
print("--- First 5 Rows of Data ---")
display(df.head())

# Handle missing values if any
print("\n--- Missing Values Count ---")
print(df.isnull().sum())
df = df.dropna()

# Display descriptive statistics
print("\n--- Summary Statistics ---")
display(df.describe())
```

## 3. Exploratory Data Analysis (EDA) & Variable Selection
To choose the single best predictor for a Simple Linear Regression model, we analyze the correlation matrix and visualize individual scatter plots with regression trends.

```python
# Calculate Pearson correlation coefficients
correlation_matrix = df.corr()
print("--- Correlation Matrix ---")
print(correlation_matrix)

print("\n--- Correlation with Target Variable (Sales) ---")
print(correlation_matrix['Sales'].sort_values(ascending=False))

# Visualize relationships across channels
fig, axes = plt.subplots(1, 3, figsize=(16, 5))
channels = ['TV', 'Radio', 'Social_Media']

for i, col in enumerate(channels):
    sns.scatterplot(data=df, x=col, y='Sales', alpha=0.6, ax=axes[i])
    sns.regplot(data=df, x=col, y='Sales', scatter=False, color='red', ax=axes[i])
    axes[i].set_title(f'{col} vs Sales')

plt.tight_layout()
plt.show()
```

### Justification for Variable Selection
Based on both numerical correlation and scatter plot indicators, **TV** shows a near-perfect linear relationship with **Sales** (correlation coefficient ~ 0.99). Therefore, **TV** is selected as the independent variable ($X$) to optimize the model's predictive power.

## 4. Build the OLS Regression Model
We use `statsmodels` to build an Ordinary Least Squares (OLS) model, adding a constant term to estimate the intercept parameter.

```python
# Assign features and target
X = df['TV']
y = df['Sales']

# Add explicit intercept term 
X_with_const = sm.add_constant(X)

# Fit the model
model = sm.OLS(y, X_with_const).fit()
print(model.summary())
```

## 5. Model Diagnostic Plots (Assumption Testing)
To ensure the statistical validity of our model, we check three critical regression assumptions using residual outputs:
1. **Linearity & Homoscedasticity:** Checked via the Residuals vs Fitted values plot.
2. **Normality:** Verified with a normal Q-Q plot and distribution histogram.

```python
residuals = model.resid
fitted_values = model.fittedvalues

fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# 1. Residuals vs Fitted Plot
sns.scatterplot(x=fitted_values, y=residuals, alpha=0.6, ax=axes[0])
axes[0].axhline(y=0, color='red', linestyle='--')
axes[0].set_xlabel('Fitted Values (Predicted Sales)')
axes[0].set_ylabel('Residuals')
axes[0].set_title('Residuals vs Fitted (Linearity/Homoscedasticity)')

# 2. Normal Q-Q Plot
stats.probplot(residuals, dist="norm", plot=axes[1])
axes[1].set_title('Normal Q-Q Plot')

# 3. Histogram of Residuals
sns.histplot(residuals, kde=True, ax=axes[2])
axes[2].set_xlabel('Residual Value')
axes[2].set_title('Histogram of Residuals (Normality)')

plt.tight_layout()
plt.show()
```

## 6. Interpretation & Marketing Recommendations

### Business Interpretation for Non-Technical Stakeholders:
- **Model Accuracy ($R^2$):** The R-squared value shows exactly what percentage of the variance in Sales can be directly attributed to TV budget choices. A high metric implies an exceptionally predictable return.
- **Return Coefficient:** The TV slope coefficient outlines exactly how many currency units Sales will increase for every single dollar added to the TV ad campaign.
- **Statistical Significance:** The p-value for the TV metric is well under 0.05, confirming that the relationship is robust and not driven by noise.

### ROI Budget Strategy:
1. **Maximize TV Budget Allocation:** Given the superior linear alignment and direct impact metrics, TV should remain the clear anchor for corporate advertising capital.
2. **Optimize Secondary Media:** Reallocate underperforming budget from Radio or Social Media to scale TV activities until saturation benchmarks are established.
